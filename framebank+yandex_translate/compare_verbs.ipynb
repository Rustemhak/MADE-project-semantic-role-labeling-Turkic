```python
import pandas as pd
import ast
```

```python
pd.set_option("max_info_columns", 5)
pd.set_option("display.max_colwidth", None)
```

```python
res_2_models_ba = pd.read_csv('data_SRL/res_2_models_ba.csv')
```

```python
res_2_models_ba.head()
```

```python
print(ast.literal_eval(res_2_models_ba.loc[0, 'srl_ba'])['verbs'][0]['verb'])
print(ast.literal_eval(res_2_models_ba.loc[0, 'result_srl_kk'])['verbs'][0]['verb'])
```

```python
for i in range(len(res_2_models_ba)):
    try:
        verb1 = ast.literal_eval(res_2_models_ba.loc[i, 'srl_ba'])['verbs'][0]['verb']
        verb2 = set()
        for j in range(len(ast.literal_eval(res_2_models_ba.loc[i, 'result_srl_kk'])['verbs'])):
            verb2.add(ast.literal_eval(res_2_models_ba.loc[i, 'result_srl_kk'])['verbs'][j]['verb'])
        if (verb1 in verb2) & (verb1 != ""):
            res_2_models_ba.loc[i, 'preds_kk_new'] = 1
        else:
            res_2_models_ba.loc[i, 'preds_kk_new'] = 0
    except:
        pass
    print(i, 'v1: ', verb1, 'v2: ', verb2, res_2_models_ba.loc[i, 'preds_kk_new'])
```
