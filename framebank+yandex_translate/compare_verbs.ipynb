```python
import string 
def drop_punkt(word):
    word=str(word)
    punc = string.punctuation
    for p in punc:
        word = word.replace(p,'')
    return word
```

```python
for i in range(len(res_2_models_ba)):
    try:
        verb1 = set()
        for j in range(len(ast.literal_eval(res_2_models_ba.loc[i, 'srl_ba'])['verbs'])):
            verb1.add(ast.literal_eval(res_2_models_ba.loc[i, 'srl_ba'])['verbs'][j]['verb'])
        verb2 = set()
        for j in range(len(ast.literal_eval(res_2_models_ba.loc[i, 'result_srl_kk'])['verbs'])):
            verb2.add(ast.literal_eval(res_2_models_ba.loc[i, 'result_srl_kk'])['verbs'][j]['verb'])
        if (verb1 in verb2) & (verb1 != ""):
            res_2_models_ba.loc[i, 'preds_kk_new'] = 1
        else:
            res_2_models_ba.loc[i, 'preds_kk_new'] = 0
        
        arguments1 = ast.literal_eval(res_2_models_ba.loc[i, 'srl_ba'])['verbs'][0]['description']
        args1 = set()
        for key, value in arguments1.items():
            if key != 'V':
                args1.add(value)
        
        args2 = set()
        
        for j in range(len(ast.literal_eval(res_2_models_ba.loc[i, 'result_srl_kk'])['verbs'])):
            arguments2 = ast.literal_eval(res_2_models_ba.loc[i, 'result_srl_kk'])['verbs'][j]['description']
            for key, value in arguments2.items():
                if key != 'V':
                    args2.add(value)
        if len(args1.intersection(args2)) != 0:
            res_2_models_ba.loc[i, 'preds_kk_args'] = 1
        else:
            res_2_models_ba.loc[i, 'preds_kk_args'] = 0
    except:
        pass
    print(i, 'v1: ', verb1, 'v2: ', verb2, res_2_models_ba.loc[i, 'preds_kk_new'], 'args1: ', args1,
          'args2: ', args2, res_2_models_ba.loc[i, 'preds_kk_args'])
```
